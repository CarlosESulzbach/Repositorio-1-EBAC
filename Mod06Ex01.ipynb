```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
bd = pd.read_csv('b1abb33c4f.csv', sep = ',')
bd.head()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
fig = plt.figure()
plt.rc('figure', figsize=(15, 10))
fig, axes = plt.subplots(2, 3)

sns.countplot(ax = axes[0, 0], x='APGAR1', data=bd)
sns.countplot(ax = axes[0, 1], x='APGAR5', data=bd)
sns.countplot(ax = axes[0, 2], x='SEXO', data=bd)
x= sns.countplot(ax = axes[1, 0], x='GESTACAO', data=bd)
sns.countplot(ax = axes[1, 1], x='ESTCIVMAE', data=bd)

x.set_xticklabels(x.get_xticklabels(), rotation=45)

plt.subplots_adjust(wspace=0.5, hspace=0.5)
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
bd_subset = bd[['APGAR5', 'CONSULTAS']]
bd_subset['CONSULTAS'].unique()
bd_subset.drop(bd[bd['CONSULTAS'] == 9].index, inplace = True)

plt.close('all')

plt.rc('figure', figsize=(15, 10))

sns.barplot(x='APGAR5', y='CONSULTAS', data=bd, ci = None)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.close('all')

fig = plt.figure()
plt.rc('figure', figsize=(15, 10))
fig, axes = plt.subplots(2, 2)

sns.countplot(ax = axes[0, 0], x='SEXO', data=bd)
x= sns.countplot(ax = axes[0, 1], x='GESTACAO', data=bd)
sns.countplot(ax = axes[1, 0], x='ESTCIVMAE', data=bd)
sns.countplot(ax = axes[1, 1], x='CONSULTAS', data=bd)

x.set_xticklabels(x.get_xticklabels(), rotation=45)

plt.subplots_adjust(wspace=0.5, hspace=0.5)
```
